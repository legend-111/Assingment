```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes
```

```python
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df.head(10)
```

```python
y = np.sin(df)
y
```

```python
z = np.cos(df)
z
```

```python
plt.plot(df,y)
plt.show()
```

```python
plt.plot(df,z)
plt.xlabel('Age')
plt.ylabel('Sex')
plt.title('Diabetes')
plt.show()
```
